```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
I = mpimg.imread('tai.jpg')
print('Img size:', I.shape)

input_img = I.reshape((I.shape[0]*I.shape[1],I.shape[2]))
print('Img size after flattening:', input_img.shape)

img_out = np.zeros_like(input_img)

k_means = KMeans(n_clusters = 4).fit(input_img)
label = k_means.predict(input_img)

print('Centers: ', k_means.cluster_centers_)
print('Label: ', label)

plt.subplot(1,4,1)
plt.imshow(I)
plt.axis('off')
plt.title('Input image')

for K in range(1, 4):
    for k in range(K):
        img_out[label == k] = k_means.cluster_centers_[k]
        output = img_out.reshape(I.shape[0],I.shape[1],I.shape[2])
    plt.subplot(1,4,int(K + 1))
    plt.imshow(output)
    plt.axis('off')
    plt.title('k = ' + str(K))

plt.show()
```
